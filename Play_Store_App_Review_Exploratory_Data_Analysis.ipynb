**"App" Column :**

````python
#unique values for "App" Column :

list(df1["App"].unique())
````

````python
# These App Names have various special character like '-','â€”',"(","|","/","â€“","(",":" which needs to be removed 
special_char = ['|',':',"(",'-',"/","â€“","(","â€”","-"]

# Defining a function for removing these extra descriptions :
def removing_extras(list1,df):
  for i in special_char:
    df["App"] = df["App"].apply(lambda x : x.split(i,1)[0] if i in x else x)

# Applying Function :
removing_extras(special_char,df1)

# Printing the Apps after removing the disturbance :
df1["App"].unique().tolist()
````

**"Rating" Column:**

````python
#unique values for "Rating" Column:

df1["Rating"].unique()
````

````python
#There's entries which are invalid : nan 

print(df1["Rating"].isnull().sum())
````

````python
#Total of 1394 peoples haven't rated the app
#Lets Remove null values 

df1.dropna(subset = ['Rating'],inplace = True)
````

**"Reviews" Column :**

````python
#unique values for "Reviews" Column :

print(df1["Reviews"].unique().tolist())
````

````python
#These Values are in string type, so we will covert it to int:
df1["Reviews"] = df1["Reviews"].apply(int)
````

**"Size" Column :**

````python
#unique values for "Size" Column :
df1["Size"].unique().tolist()
````

````python
#Review are of string data type which include integer values and M and K at the end :
#Also some entries says "Varies with device" which also need to be coverted into nan:

df1["Size"] = df1["Size"].apply(lambda x : str(x).replace("Varies with device","nan") if "Varies with device" in str(x) else x)
df1["Size"] = df1["Size"].apply(lambda x : str(x).replace("M","")if 'M' in str(x) else x)
df1["Size"] = df1["Size"].apply(lambda x : float(str(x).replace("k",""))/1000 if "k" in str(x) else x)
````

````python
#So we just completed handling M,K and Varies with device
#Now we will convert this data type into float:

df1["Size"] = df1["Size"].apply(float)
````

````python
#Lets check for null values

df1["Size"].isnull().sum()
````

````python
#Filling the null values

df1.fillna(df1["Size"].mean(),inplace = True)
````

**"Installs" Column :**

````python
#unique values for "Installs" Column :
df1["Installs"].unique().tolist()
````

````python
#Values inside Install column contain "," and "+" 
#first we will remove these two:

df1["Installs"] = df1["Installs"].apply(lambda x : x.replace(",","") if "," in x else x)
df1["Installs"] = df1["Installs"].apply(lambda x : x.replace("+","") if "+" in x else x)
````
